## Criando e excluindo variáveis

```python
import pandas as pd
dados = pd.read_csv('aluguel_residencial.csv', sep = ';')
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)

casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')

dados
```

```python
# Excluindo
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
dados_aux.head(10)
```

```python
del dados_aux['Valor Bruto']
dados_aux.head(10)
```

```python
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
dados.head(10)
```

## Gráficos

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
aluguel = pd.read_csv('aluguel_residencial.csv', sep = ';')
aluguel.head(5)
```

```python
# Gráfico de barras para a contagem de imóveis por tipo
plt.figure(figsize=(10, 6))  # Tamanho da figura
tipo_counts = aluguel['Tipo'].value_counts()
tipo_counts.plot(kind='bar', color='green')  # Cor para verde
plt.title('Contagem de Imóveis por Tipo')
plt.xlabel('Tipo de Imóvel')
plt.ylabel('Contagem')
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.tight_layout()  # Melhora o espaçamento para evitar cortes no gráfico

# Gráfico de barras para a contagem de imóveis por bairro
plt.figure(figsize=(12, 6))  # Tamanho da figura
bairro_counts = aluguel['Bairro'].value_counts().head(10)  # Ajustar a quantidade de bairros a serem exibidos
bairro_counts.plot(kind='bar', color='purple')  # Cor para roxo
plt.title('Contagem de Imóveis por Bairro (Top 10)')
plt.xlabel('Bairro')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.tight_layout()

# Exiba os gráficos
plt.show()
```

```python
# Escolha os tipos de imóveis ou bairros a serem incluídos no gráfico
tipos_selecionados = aluguel['Tipo'].value_counts().head(10).index.tolist()  # Top 10 tipos de imóveis
bairros_selecionados = aluguel['Bairro'].value_counts().head(10).index.tolist()  # Top 10 bairros

# Gráfico de barras para a média de valores por tipo de imóvel
plt.figure(figsize=(10, 6))  # Tamanho da figura
media_por_tipo = aluguel[aluguel['Tipo'].isin(tipos_selecionados)].groupby('Tipo')['Valor'].mean()
media_por_tipo.plot(kind='bar', color='green')
plt.title('Média de Valor por Tipo de Imóvel')
plt.xlabel('Tipo de Imóvel')
plt.ylabel('Média de Valor')
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.tight_layout()  # Melhora o espaçamento para evitar cortes no gráfico

# Gráfico de barras para a média de valores por bairro
plt.figure(figsize=(12, 6))  # Tamanho da figura
media_por_bairro = aluguel[aluguel['Bairro'].isin(bairros_selecionados)].groupby('Bairro')['Valor'].mean()
media_por_bairro.plot(kind='bar', color='purple')
plt.title('Média de Valor por Bairro')
plt.xlabel('Bairro')
plt.ylabel('Média de Valor')
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.tight_layout()

# Exiba os gráficos
plt.show()
```

```python
# Contagem de cada tipo de imóvel
contagem_tipo = aluguel['Tipo'].value_counts()

# Paleta de cores personalizada para cada tipo de imóvel
cores_personalizadas = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#c2c2f0', '#ffb3e6', '#c2f0c2']

# Gráfico de Pizza com cores personalizadas
plt.figure(figsize=(8, 8))  # Ajuste o tamanho conforme necessário
plt.pie(contagem_tipo, labels=contagem_tipo.index, autopct='%1.1f%%', colors=cores_personalizadas)
plt.title('Proporção de Tipos de Imóveis')
plt.axis('equal')  # Mantém o aspecto de um círculo

# Exiba o Gráfico de Pizza
plt.show()
```

```python
# Escolha as variáveis numéricas de interesse
variaveis_numericas = ['Quartos', 'Suites', 'Area']

# Subplot para cada variável numérica
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 5))  # Ajuste o tamanho conforme necessário

# Configurações específicas para cada variável
configuracoes = {
    'Quartos': {'titulo': 'Histograma de Quartos', 'limite_x': (0, 6), 'cor': 'purple'},
    'Suites': {'titulo': 'Histograma de Suites', 'limite_x': (0, 6), 'cor': 'blue'},
    'Area': {'titulo': 'Histograma de Área', 'limite_x': (0, 1000), 'cor': 'green'}
}

# Histograma para cada variável numérica
for i, variavel in enumerate(variaveis_numericas):
    ax = axes[i]

    # Plote o histograma na cor especificada
    ax.hist(aluguel[variavel], bins=20, color=configuracoes[variavel]['cor'], alpha=0.7)

    # Adicione uma linha de tendência preta
    media = aluguel[variavel].mean()
    ax.axvline(media, color='black', linestyle='dashed', linewidth=2)

    ax.set_title(configuracoes[variavel]['titulo'])
    ax.set_xlabel(variavel)
    ax.set_ylabel('Frequência')
    ax.set_xlim(configuracoes[variavel]['limite_x'])
    ax.grid(True)

# Ajuste o layout
plt.tight_layout()

# Exiba os histogramas
plt.show()
```
